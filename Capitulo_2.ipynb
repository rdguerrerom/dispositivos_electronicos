#Aplicaciones de la física de dispositivos electrónicos

##Transistores

Un transistor es un dispositivo semiconductor que se utiliza para amplificar o cambiar señales electrónicas y energía eléctrica. Los transistores son uno de los componentes básicos de la electrónica moderna.  Un transistor está compuesto de material semiconductor generalmente con al menos tres terminales para la conexión a un circuito externo. Un voltaje o corriente aplicada a un par de terminales del transistor controla la corriente a través de otro par de terminales. Debido a que la potencia controlada (salida) puede ser mayor que la potencia controladora (entrada), un transistor puede amplificar una señal. Hoy en día, algunos transistores se empaquetan individualmente, pero muchos más se encuentran integrados en circuitos integrados.

###Transistor de unión bipolar (BJT)

4-1.png

El BJT está construido con tres regiones semiconductoras dopadas separadas por dos uniones $pn$, como se muestra en la estructura plana epitaxial de la figura 4-1 (a). Las tres regiones se denominan emisor(E), base(B) y colector(C). Las representaciones físicas de los dos tipos de BJT se muestran en la Figura 4-1 (b) y (c). Un tipo consta de dos  regiones $n$ separadas por una  región $p$ ($npn$), y el otro tipo consta de dos  regiones $p$ separadas por una  región $n$ ($pnp$). El término bipolar se refiere al uso de huecos y electrones como portadores de corriente en la estructura del transistor.

La unión $pn$ que une la región base y la región emisora se denomina *unión base-emisor*. La unión $pn$ que une la región base y la región colectora se denomina *unión base-colector*, como se indica en la figura 4-1 (b). Un cable se conecta a cada una de las tres regiones, como se muestra. Estos cables están etiquetados como E, B y C para emisor, base y colector, respectivamente. La región de la base está ligeramente dopada y es muy delgada en comparación con el emisor fuertemente dopado y las regiones colectoras moderadamente dopadas. Debido a esta diferencia en los niveles de dopaje, **el emisor y el recolector no son intercambiables**.

####Operación básica de un BJT

Para que un BJT funcione correctamente como amplificador, las dos uniones pn deben estar polarizadas correctamente con voltajes de CC externos. En esta sección, usamos principalmente el transistor $npn$ como ilustración. El funcionamiento del pnp es el mismo que el del $npn$, excepto que las funciones de los electrones y huecos, las polaridades del voltaje de polarización y las direcciones de la corriente se invierten.

#####Operación

Para comprender cómo funciona un transistor, examinemos qué sucede dentro de la estructura $npn$. La región emisora de tipo $n$ fuertemente dopada tiene una densidad muy alta de electrones de banda de conducción (libres), como se indica en la figura 4-4. Estos electrones libres se difunden fácilmente a través de la *unión BE* de base directa hacia la región de la base de tipo $p$ ligeramente dopada y muy delgada, como lo indica la flecha ancha. La base tiene una baja densidad de huecos, que son los portadores mayoritarios, como lo representan los círculos blancos. Un pequeño porcentaje del número total de electrones libres inyectados en la región de la base se recombina con los huecos y se mueve como electrones de valencia a través de la región de la base y hacia la región del emisor como la corriente del hueco, indicado por las flechas rojas.

4-4.png

Cuando los electrones que se han recombinado con huecos a meedida que los electrones de valencia abandonan la estructura cristalina de la base, se convierten en electrones libres de la base metálica y producen la corriente de base externa. La mayoría de los electrones libres que han entrado en la base no se recombinan con los huecos porque la base es muy delgada. A medida que los electrones libres se mueven hacia la *unión BC* con polarización inversa, son arrastrados hacia la región del colector por la atracción de la tensión de alimentación positiva del colector. Los electrones libres se mueven a través de la región del colector hacia el circuito externo y luego regresan a la región del emisor junto con la corriente base, como se indica. La corriente del emisor es ligeramente mayor que la corriente del colector debido a la pequeña corriente de base que se separa de la corriente total inyectada en la región de la base desde el emisor.

Cuando un transistor está conectado a voltajes de polarización de CC, como se muestra en la Figura 4-6 para los tipos $npn$ y $pnp$, $V_{BB}$ polariza hacia adelante la unión base-emisor y $V_{CC}$ polariza inversamente la unión base-colector. #####Parámetros $\beta_{DC}$ y $\alpha_{DC}$ 

La ganancia de corriente de CC de un transistor es la relación entre la corriente del colector de CC $ \left (I_ {C} \right) $ y la corriente de CC de base $ \left (I _ {\mathrm {B}} \right) $ y es designado dc beta $ \left (\beta _ {\mathrm {DC}} \right) $.
$$
\boldsymbol{\beta}_{\mathrm{DC}}=\frac{\boldsymbol{I}_{\mathrm{C}}}{\boldsymbol{I}_{\mathrm{B}}} \tag{4.2}
$$
Los valores típicos de $ \beta _ {\mathrm {DC}} $ oscilan entre menos de 20 y 200 o más. $ \beta _ {\mathrm {DC}} $ generalmente se designa como un parámetro $ (h) $ híbrido equivalente, $ h _ {\mathrm {FE}} $, en las hojas de datos del transistor. Los parámetros $ h $ se tratan en el Capítulo 6. Todo lo que necesitas saber ahora es que
$$
h_{\mathrm{FE}}=\beta_{\mathrm{DC}}
$$
La relación entre la corriente del colector de cd $ \left (I_ {C} \right) $ y la corriente del emisor de cd $ \left (I_ {E} \right) $ es el alfa de DC $ \left (\alpha _ {\mathrm { DC}} \right). $ El alfa es un parámetro menos utilizado que el beta en los circuitos de transistores.
$$
\alpha_{\mathrm{DC}}=\frac{I_{\mathrm{C}}}{I_{\mathrm{E}}}
$$
Normalmente, los valores de $ \alpha _ {\mathrm {DC}} $ oscilan entre $ 0,95 $ y $ 0,99 $ o más, pero $ \alpha _ {\mathrm {DC}} $ siempre es menor que 1. La razón es que $ I _ {\mathrm {C}} $ siempre es ligeramente menor que $ I _ {\mathrm {E}} $ por la cantidad de $ I _ {\mathrm {B}}. $ Por ejemplo, si $ I_ {\mathrm {E}} = 100 \mathrm {~ mA} $ y $ I _ {\mathrm {B}} = 1 \mathrm {~ mA} $, luego $ I _ {\mathrm {C}} = 99 \mathrm {~ mA} $ y $ \alpha _ {\mathrm {DC}} = 0,99 $

######Ejemplo

Determine la ganancia de corriente continua $ \beta _ {\mathrm {DC}} $ y la corriente del emisor $ I _ {\mathrm {E}} $ para un transistor donde $ I _ {\mathrm {B}} = 50 \mu \mathrm { A} $ y $ I _ {\mathrm {C}} = 3.65 \mathrm {~ mA} $
Solución
$$
\begin{aligned}
\beta_{\mathrm{DC}} &=\frac{I_{\mathrm{C}}}{I_{\mathrm{B}}}=\frac{3.65 \mathrm{~mA}}{50 \mu \mathrm{A}}=73 \\
I_{\mathrm{E}} &=I_{\mathrm{C}}+I_{\mathrm{B}}=3.65 \mathrm{~mA}+50 \mu \mathrm{A}=3.70 \mathrm{~m} \mathrm{~A}
\end{aligned}
$$

#####Análisis de circuito basico para un BJT

Considere la configuración básica del circuito de polarización del transistor en la Figura 4-8. Se pueden identificar tres corrientes de CC de transistor y tres voltajes de CC.
* $ I _ {\mathrm {B}} $: corriente de base de CC
* $ I _ {\mathrm {E}}: $ dc emisor de corriente
* $ I _ {\mathrm {C}}: $ dc colector actual
* $ V _ {\mathrm {BE}}: \mathrm {dc} $ voltaje en la base con respecto al emisor
* $ V _ {\mathrm {CB}}: voltaje de $ dc en el colector con respecto a la base
* $ V _ {\mathrm {CE}}: voltaje de $ dc en el colector con respecto al emisor


4-8.png

La fuente de voltaje de polarización de base, $ V _ {\ mathrm {BB}} $, polariza hacia adelante la unión base-emisor, y la fuente de voltaje de polarización de colector, $ V _ {\ mathrm {CC}} $, polariza de forma inversa el Unión base-colector. Cuando la unión base-emisor está polarizada hacia adelante, es como un diodo polarizado hacia adelante y tiene una caída nominal de voltaje directo de
$$
V_{\mathrm{BE}} \cong 0.7 \mathbf{V} \tag{4.3}
$$
Aunque en un transistor real $ V _ {\ mathrm {BE}} $ puede ser tan alto como $ 0.9 \ mathrm {~ V} $ y depende de la corriente, usaremos $ 0.7 \ mathrm {~ V} $ a lo largo de este texto en orden para simplificar el análisis de los conceptos básicos. Tenga en cuenta que la característica de la unión base-emisor es la misma que la de una curva de diodo normal como la de la Figura $ 2-12. $

Dado que el emisor está en el suelo $ (0 \mathrm {~ V}) $, según la ley de voltaje de Kirchhoff, el voltaje en $ R _ {\mathrm {B}} $ es
$$
V_{R _{\mathrm {B}}} = V _ {\mathrm {BB}} - V _ {\mathrm {BE}}
$$
Además, según la ley de Ohm,
$$
V_{R_{\mathrm{B}}}=I_{\mathrm{B}} R_{\mathrm{B}}
$$
Sustituyendo los rendimientos de $ V_{R _{\mathrm {B}}} $
$$
I_{\mathrm{B}} R_{\mathrm{B}}=V_{\mathrm{BB}}-V_{\mathrm{BE}} \tag{4.4}
$$
Resolviendo para $ I_{\mathrm {B}} $,
$$
I_{\mathrm{B}}=\frac{\boldsymbol{V}_{\mathrm{BB}}-\boldsymbol{V}_{\mathrm{BE}}}{\boldsymbol{R}_{\mathrm{B}}}
$$
El voltaje en el colector con respecto al emisor conectado a tierra es
$$
V_{\mathrm{CE}}=V_{\mathrm{CC}}-V_{R_{C}}
$$
Dado que la caída en $ R_ {C} $ es
$$
V_{R_{\mathrm{C}}}=I_{\mathrm{C}} R_{\mathrm{C}}
$$
el voltaje en el colector con respecto al emisor se puede escribir como
$$
V_{\mathrm{CE}}=V_{\mathrm{CC}}-I_{\mathrm{C}} R_{\mathrm{C}}\tag{4.5}
$$
donde $ I _ {\mathrm {C}} = \beta _ {\mathrm {DC}} I _ {\mathrm {B}} $.
El voltaje a través de la unión colector-base con polarización inversa es
$$
V_{\mathrm{CB}}=V_{\mathrm{CE}}-V_{\mathrm{BE}} \tag{4.6}
$$

impedancia a tierra y, por lo tanto, $ I_ {C} $ será aproximadamente cero y $ V _ {\mathrm {CE}} $ está cerca de $ 0 \mathrm {~ V} $. Cuando ambas uniones están polarizadas hacia adelante, el transistor está en la región de saturación de su operación. La saturación es el estado de un BJT en el que la corriente del colector ha alcanzado un máximo y es independiente de la corriente base.

A medida que aumenta $ V _ {\mathrm {CC}} $, $ V _ {\mathrm {CE}} $ aumenta a medida que aumenta la corriente del colector. Esto está indicado por la porción de la curva característica entre los puntos $ A $ y $ B $ en la Figura $ 4-10 (b). I_ {C} $ aumenta a medida que $ V _ {\mathrm {CC}} $ aumenta porque $ V _ {\mathrm {CE}} $ permanece por debajo de $ 0.7 \mathrm {~ V} $ debido al colector base con polarización directa unión.

Idealmente, cuando $ V _ {\mathrm {CE}} $ excede $ 0.7 \mathrm {~ V} $, la unión base-colector se polariza inversamente y el transistor entra en la región activa o lineal de su operación. Una vez que la unión del colector base tiene polarización inversa, $ I _ {\mathrm {C}} $ se nivela y permanece esencialmente constante para un valor dado de $ I _ {\mathrm {B}} $ como $ V _ {\mathrm {CE}} $ sigue aumentando. En realidad, $ I _ {\ mathrm {C}} $ aumenta muy levemente a medida que $ V _ {\mathrm {CE}} $ aumenta debido al ensanchamiento de la región de agotamiento del colector base. Esto da como resultado menos agujeros para la recombinación en la región base, lo que efectivamente causa un ligero aumento en $ \beta _ {\mathrm {DC}} $. Esto se muestra en la parte de la curva característica entre los puntos $ B $ y $ C $ en la Figura $ 4-10 (\mathrm {~ b}) $. Para esta parte de la curva característica, el valor de $ I _ {\mathrm {C}} $ está determinado solo por la relación expresada como $ I _ {\mathrm {C}} = \beta _ {\mathrm {DC}} I_ { \mathrm {B}} $.

Cuando $ V _ {\mathrm {CE}} $ alcanza un voltaje suficientemente alto, la unión base-colector con polarización inversa entra en ruptura; y la corriente del colector aumenta rápidamente como lo indica la parte de la curva a la derecha del punto $ C $ en la figura $ 4-10 (b) $. Nunca se debe operar un transistor en esta región de ruptura.

#####Saturación

4-14.png

Cuando la unión base-emisor se polariza hacia adelante y la corriente base aumenta, la corriente del colector también aumenta $ \left (I _ {\mathrm {C}} = \beta _ {\mathrm {DC}} I _ {\mathrm {B }} \right) $ y $ V _ {\mathrm {CE}} $ disminuye como resultado de una mayor caída en la resistencia del colector $ \left (V _ {\mathrm {CE}} = V _ {\mathrm {CC}} - I _ {\mathrm {C}} R _ {\mathrm {C}} \right). $ Esto se ilustra en la Figura $ 4-14. $ Cuando $ V _ {\mathrm {CE}} $ alcanza su valor de saturación, $ V_ { \mathrm {CE} (\mathrm {sat})} $, la unión base-colector se vuelve sesgada hacia adelante y $ I _ {\mathrm {C}} $ no puede aumentar más incluso con un aumento continuo en $ I _ {\mathrm {B}} $. En el punto de saturación, la relación $ I _ {\mathrm {C}} = \beta _ {\mathrm {DC}} I _ {\mathrm {B}} $ ya no es válida. $ V _ {\mathrm {CE} (\mathrm {sat})} $ para un transistor ocurre en algún lugar por debajo de la rodilla de las curvas del colector, y generalmente es solo unas pocas décimas de voltio.

#####Valoraciones máximas

Un BJT, como cualquier otro dispositivo electrónico, tiene limitaciones en su funcionamiento. Estas limitaciones se establecen en forma de clasificaciones máximas y normalmente se especifican en la hoja de datos del fabricante. Normalmente, las clasificaciones máximas se dan para voltaje de colector a base, voltaje de colector a emisor, voltaje de emisor a base, corriente de colector y disipación de potencia.
El producto de $ V_{\mathrm {CE}} $ y $ I_{\mathrm {C}} $ no debe exceder la disipación de potencia máxima, $ P _ {\mathrm {D} (\max)} $. (Las hojas de características pueden mostrar esto simplemente como $ P_{\mathrm {D}} $ en una columna denominada Calificaciones máximas). Tanto $ V _ {\mathrm {CE}} $ y $ I _ {\mathrm {C}} $ no pueden ser máximo al mismo tiempo. Si $ V _ {\mathrm {CE}} $ es el máximo, $ I _ {\mathrm {C}} $ se puede calcular como
$$
I_{\mathrm{C}}=\frac{P_{\mathrm{D}(\max )}}{V_{\mathrm{CE}}}
$$
Si $ I _ {\mathrm {C}} $ es el máximo, $ V_{\mathrm {CE}} $ se puede calcular reordenando la ecuación anterior de la siguiente manera:
$$
V_{\mathrm{CE}}=\frac{P_{\mathrm{D}(\max )}}{I_{\mathrm{C}}}
$$
Para cualquier transistor dado, se puede trazar una curva de disipación de potencia máxima en las curvas características del colector, como se muestra en la Figura 4-18 (a). Estos valores se tabulan en la Figura 4-18 (b). Suponga que $ P_{\mathrm {D} (\max)} $ es $ 500 \mathrm{~ mW}, V _ {\mathrm {CE} (\max)} $ es $ 20 \mathrm {~ V} $ y $ I_ {\mathrm {C} (\max)} $ es $ 50 \mathrm {~ mA} $. La curva muestra que este transistor en particular no se puede operar en la parte sombreada del gráfico. $ I _ {\mathrm {C} (\max)} $ es la calificación límite entre los puntos $ A $ y $ B, P _ {\mathrm {D} (\max)} $ es la calificación límite entre los puntos $ B $ y $ C $, y $ V _ {\mathrm {CE} (\max)} $ es la calificación límite entre los puntos $ C $ y $ D $.

######Ejemplo

Cierto transistor debe operarse con $ V _ {\mathrm {CE}} = 6 \mathrm {~ V} $. Si su potencia nominal máxima es $ 250 \mathrm {~ mW} $, ¿cuál es la mayor corriente de colector que puede manejar?

**Solution**
$$
I_{\mathrm{C}}=\frac{P_{\mathrm{D}(\max )}}{V_{\mathrm{CE}}}=\frac{250 \mathrm{~mW}}{6 \mathrm{~V}}=41.7 \mathrm{~mA}
$$
Esta  es la corriente para este valor particular de  $V_{\mathrm{CE}}$. l transistor puede manejar mas corriente en el colector si s reduce $V_{\mathrm{CE}}$, siempre y cuando $P_{\mathrm{D}(\max )}$ y $I_{\mathrm{C}(\max )}$ no se excedan.
